From reading .txt file attached to dataset we know that instead NaN we have unknown, let's replace all unknown with NaN <br>
Also said that 'duration' should be discarded if the intention is to have a realistic predictive model, so it will be removed

```python
data.replace('unknown', np.NaN, inplace=True)
data.isnull().sum()
```

Let's count the percentage of missing data

```python
#missing data
total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum() / data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
data.drop(['default', 'duration'], axis=1, inplace=True)
```

Let's try to find columns that have all unique labels

```python
(data.nunique() / len(data)).sort_values(ascending = False)
```

Inspecting features I've found feature that in my opinion is not necessary: 

pdays: number of days that passed by after the client was last contacted from a previous campaign

```python
data.pdays.value_counts()
```

We have 999 in there (means client was not previously contacted), but due to that it's mostly 999, i'd like to remove it (of course we could replace all 999 with 0 so that would make more sense, but in my opinion it's better to remove it completely), also it is kind of replaceble with 'previous' <br>
previous: number of contacts performed before this campaign and for this client

```python
data.drop(['pdays'], axis=1, inplace=True)
```

Now we will divide features based on their type

```python
numeric_features = data.select_dtypes('number').columns
categorical_features = data.select_dtypes(np.object).columns
```

Now we will fill missing numerical values:

```python
data[numeric_features].describe().T
```

```python
# sns.pairplot(data[numeric_features])
```

Since almost all of features are not symmetric we will impute with 'median'

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median', verbose=True)

data[numeric_features] = imputer.fit_transform(data[numeric_features])

data[numeric_features].describe().T
```

Now we will fill missing categorical values:

```python
imputer_freq = SimpleImputer(missing_values=np.nan, strategy='most_frequent', verbose=True)

data[categorical_features] = imputer_freq.fit_transform(data[categorical_features])

data[categorical_features].isnull().sum()
```

Now let's scale our numeric data

```python
data[numeric_features].describe()
```

Since our data is not normally distributed (at least 'previous') we will use MinMaxScaler:

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

print(data[numeric_features].head())

data[numeric_features] = min_max_scaler.fit_transform(data[numeric_features])

print(data[numeric_features])
```

Let's check if scaled it right:

```python
data[numeric_features].max()
```

```python
data[numeric_features].min()
```
