$ a = 0.9, b = 0.2, c = 0.6, \eta = 1 $

```python
import numpy as np 
import matplotlib.pyplot as plt

a = 0.9
b = 0.2
c = 0.6
eta = 1

r = np.array([a, b])
R = np.array([[1, c], 
              [c, 1]]) # type: ignore
sigma = 0.6
```

```python
def cost_f(sigma, r, w, R):
    return 0.5*sigma**2 - np.dot(r.T, w) + 0.5 * np.dot(np.dot(w.T, R), w)

def f(x, y):
    return 0.5*sigma**2 - a*x - b*y+ 0.5*(x**2 + 2*c*x*y + y**2)
```

```python
def gradient(N, eta, r, R, sigma):
    w_result = []
    cost_result = []
    w_0 = np.array([0.6, 0.6])
    w_result.append(w_0)
    cost_result.append(cost_f(sigma, r, w_0, R))
    for i in range(N):
        w_result.append(w_result[i] + eta * (r - np.dot(R, w_result[i])))
        cost_result.append(cost_f(sigma, r, w_result[i], R))
        
    return w_result, cost_result
```

```python
w, cost = gradient(200, eta, r, R, sigma)
print(f'last w: {w[-1]}, last cost: {cost[-1]}')
```

```python
# a = 0.9
# b = 0.2
# c = 0.6
import numpy as np 

def gradient(N, eta, r, R, sigma):
    w_result = []
    cost_result = []
    w_0 = np.array([0.5, 0.5])
    w_result.append(w_0)
    cost_result.append(cost_f(sigma, r, w_0, R))
    for i in range(N):
        print(f'{w_result[i]} + {eta} * {r} - R = {R}, w = {w_result[i]}{np.dot(R, w_result[i])}')
        w_result.append(w_result[i] + eta * (r - np.dot(R, w_result[i])))
        cost_result.append(cost_f(sigma, r, w_result[i], R))
        
    return w_result, cost_result

r = np.array([0.8, 0.4])
R = np.array([[1, 0.7], 
              [0.7, 1]]) # type: ignore
sigma = 1
eta = 1.4

w, cost = gradient(2, eta, r, R, sigma)
print(f'last w: {w[-1]}, last cost: {cost[-1]}')
```
