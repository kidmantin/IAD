```python
import numpy as np 

T_test = np.array([[5, 1, 6, 7], #   
                  [7, 5, 9, 8],
                  [3, 9, 6, 5],
                  [2, 3, 7, 1]])

N = np.shape(T_test)[0] 

mnozh, counts = np.unique(T_test, return_counts=True) #    T,
#     p_1 

y_test = np.array([1, 2, 3, 2])
S_mnozh, S_counts = np.unique(y_test, return_counts=True)
```

```python
def findC_h(i, T):
    setQ_h = []
    for j in range(np.shape(T)[1]):
        setQ_h.append(T[j, i])
    return setQ_h
```

```python
def djinni(t_i):
    temp = 0
    for s in S_counts:
        temp += (s/np.size(t_i))**2
    return (1 - temp)
```

```python
def findingT_i(T, c_h, col):
    T_i = []
    for i in range(np.shape(T)[0]):
        if c_h == T[i, col]:
            T_i.append(T[i])
    return T_i
```

```python
abc = []
t_i = []
G = 0

for col in range(np.shape(T_test)[1]):
    c_h_list = findC_h(col, T_test)
    for c_h in c_h_list:
        t_i_list = findingT_i(T_test, c_h, col)
        for t_i in t_i_list:
            G += np.size(t_i)/N*djinni(t_i)
    abc.append(G)
    np.delete(c_h, 0)

print(np.min(abc))
```
