```python
import numpy as np 

T_test = np.array([[1, 5, 6, 7], #створення тестових значень 
                  [7, 5, 9, 8],
                  [3, 9, 6, 5],
                  [2, 3, 7, 1]])

mnozh, p_1 = np.unique(T_test, return_counts=True) #знаходження унікальних значень масиву T,
# та знаходження значень для p_1 

y_test = np.array([1, 2, 3, 2])
S_test = np.unique(y_test)

forp_2 = np.zeros(mnozh.size)
output = np.zeros(mnozh.size)

def p_2(y:"np.ndarray", S_j:"int", T:"np.ndarray", c_h:"int") -> "int": 
    count = 0
    for i in range(y_test.size): 
        if ((y[i] == S_j) and (c_h in T[i])):
            count += 1                      # кількість прикладів у яких є c_h і які належать класу s_j
    return count

def p_3(y:"np.ndarray", S:"int", T:"np.ndarray", c_h:"int") -> "int": 
    count = 0
    for i in range(y_test.size):
        if ((y[i] != S) and (c_h in T[i])):
            count += 1                      # кількість прикладів у яких є c_h і які не належать класу s_j
    return count

for s in S_test:
    for i in range(mnozh.size):
        forp_2[i] = p_2(y_test, s, T_test, mnozh[i])

s_zvezd = np.argmax(forp_2)

for s in S_test:
    for i in range(mnozh.size):
        output[i] = p_3(y_test, s, T_test, s_zvezd)/p_1[i]

print("Значение признака: ", np.argmin(output))
```
