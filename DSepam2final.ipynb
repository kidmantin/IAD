Encoding was done right

Now let's make a pipeline: 

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import VarianceThreshold

num_pipe = Pipeline([
    ('imputer', SimpleImputer(missing_values=np.nan, strategy='median', verbose=True)),
    ('scaler', MinMaxScaler()),
    ('variance_trsh', VarianceThreshold(threshold=0.1))
])

cat_pipe = Pipeline([
    ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent', verbose=True)),
    ('ohe', OneHotEncoder(sparse=False)), 
    ('variance_trsh', VarianceThreshold(threshold=.97 * (1 - .97)))
])

pipe = ColumnTransformer(transformers=[
    ('cat', cat_pipe, categorical_features),
    ('num', num_pipe, numeric_features)
])
```

```python
X = pipe.fit_transform(data)
```

```python
X.shape
```

As we can see we got only 38 features, let's check for NaN just in case:

```python
pd.DataFrame(X).isnull().sum()
```

Dimensionality Reduction:

PCA:

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_X = pca.fit_transform(X)
```

```python
pca_df = pd.DataFrame.from_records(data=pca_X[:,:2],columns=["x","y"])
pca_df["target"] = y_encoded
pca_df
```

```python
sns.relplot(x="x", y="y", data=pca_df, hue="target", palette='coolwarm')
plt.title('PCA')
```

```python
targetplots(pca_df)
```

```python
print(f'pca.explained_variance_ratio_: {pca.explained_variance_ratio_}, sum of it: {pca.explained_variance_ratio_.sum()}')
```

We got 4 clusters, difference between 'target == 1' and 'target == 0' is not visible to the eye and a low explained_variance

t-SNE:

```python
tsne_df = pd.DataFrame.from_records(data=tsne_X[:,:2],columns=["x","y"])
tsne_df["target"] = y_encoded
tsne_df
```

```python
sns.relplot(x="x", y="y", data=tsne_df, hue="target", palette='coolwarm')
plt.title('t-SNE')
```
