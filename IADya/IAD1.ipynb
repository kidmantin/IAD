```python
import numpy as np 

T_test = np.array([[1, 5, 6, 7], #   
                  [7, 5, 9, 8],
                  [3, 9, 6, 5],
                  [2, 3, 7, 1]])

mnozh, counts = np.unique(T_test, return_counts=True) #    T,
#     p_1 

y_test = np.array([1, 2, 3, 2])
S_test = np.unique(y_test)

output = np.zeros(mnozh.size)

def p_2(y:"np.ndarray", S:"int", T:"np.ndarray", c_h:"int") -> "int": 
    count = 0
    for i in range(y_test.size):
        if ((y[i] == S) and (c_h in T[i])):
            count += 1                      #      c_h     s_j
    return count

for s in S_test:
    for i in range(mnozh.size):
        output[i] = p_2(y_test, s, T_test, mnozh[i])/counts[i]
    print("for s =", s, "C*_h = ", np.argmax(output))
```
