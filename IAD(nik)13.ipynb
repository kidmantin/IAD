```python
import numpy as np
import matplotlib.pyplot as plt 
import random
import math

random.seed(17)
N = 200 
```

```python
center_1 = [0, 10]
center_2 = [15, 20]

radius = 5

def clusterization(center, radius, N):
    x = [] #иксы
    y = [] #игрики
    u = [] #матрица разбития 
    for i in range(N):
        x.append(random.uniform(center[0] - radius, center[0] + radius))
        y.append(random.uniform(center[1] - radius, center[1] + radius))
        u.append(1)
    return x, y, u


def countPC(matrix):
    PC = 0
    for i in matrix:
        for j in i:
            PC += j**2
    return PC/np.shape(matrix)[1]

x1, y1, u1 = clusterization(center_1, radius, N)
x2, y2, u2 = clusterization(center_2, radius, N)

plt.scatter(x1, y1, c='#ff7f0e')
plt.scatter(x2, y2, c='#9467bd')
plt.show()

u_1 = np.append(np.array(u1), np.ones(len(u1)) - np.array(u2))
u_2 = np.append(np.ones(len(u2)) - np.array(u1), np.array(u2))

matrix = np.vstack((u_1, u_2))

countPC(matrix)
```

```python

center_1 = [0, 10]
center_2 = [15, 20]

radius = 5

def clusterization(center, radius, N):
    x = [] #иксы
    y = [] #игрики
    u = [] #матрица разбития 
    for i in range(N):
        x.append(random.uniform(center[0] - radius, center[0] + radius))
        y.append(random.uniform(center[1] - radius, center[1] + radius))
        distance = math.sqrt((x[i] - center[0])**2 + (y[i] - center[1])**2)
        u.append(1 - distance/radius)
    return x, y, u


def countPC(matrix):
    PC = 0
    for i in matrix:
        for j in i:
            PC += j**2
    return PC/np.shape(matrix)[1]

x1, y1, u1 = clusterization(center_1, radius, N)
x2, y2, u2 = clusterization(center_2, radius, N)

plt.scatter(x1, y1, c='#ff7f0e')
plt.scatter(x2, y2, c='#9467bd')
plt.show()

u_1 = np.append(np.array(u1), np.ones(len(u1)) - np.array(u2))
u_2 = np.append(np.ones(len(u2)) - np.array(u1), np.array(u2))

matrix = np.vstack((u_1, u_2))

countPC(matrix)
```
