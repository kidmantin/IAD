```python
import numpy as np
import matplotlib.pyplot as plt 
import networkx as nx 
```

```python
def show_graph(graph, edge_weights):
    plt.figure(figsize=(8, 6)) 
    pos = nx.circular_layout(graph)

    # nodes
    nx.draw_networkx_nodes(graph, pos, node_size=700)

    # edges
    nx.draw_networkx_edges(graph, pos, width=3)

    # node labels
    nx.draw_networkx_labels(graph, pos, font_size=20)
    # edge weight labels
    nx.draw_networkx_edge_labels(graph, pos, edge_weights, font_size=20)

    # ax = plt.gca()
    # ax.margins(0.08)
    # plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
edges = np.array([(1, 6, 1), # 1 -  , 6 -  , 1 - 
                  (1, 5, 6),
                  (1, 4, 4),
                  (1, 3, 2),
                  (2, 6, 4),
                  (2, 4, 2),
                  (2, 3, 5),
                  (4, 6, 6),
                  (5, 6, 2)])

G = nx.Graph()

G.add_weighted_edges_from(edges)

edge_weights = nx.get_edge_attributes(G, "weight")

show_graph(G, edge_weights)
```

```python
def clusterization(graph, sigma):
    edge_to_delete =[] 
    edge_weights = nx.get_edge_attributes(graph, "weight") # 
    for nodes in edge_weights:
        if edge_weights[nodes] > sigma:
            edge_to_delete.append(nodes)

    graph.remove_edges_from(edge_to_delete) # 
    
    edge_weights = nx.get_edge_attributes(graph, "weight") #  
    return graph, edge_weights
```

```python
G_clust, edge_weights_clust = clusterization(G, sigma=3)
```

```python
show_graph(G_clust, edge_weights_clust)
```
